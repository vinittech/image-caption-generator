## Visualise Loss

We plot the training loss history capture per epoch.WE stored history per epoch so we first combine them together.

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Jan 12 15:41:38 2020

@author: dhamuk
"""

import pickle
import matplotlib.pyplot as plt   
    
history=[] 
###load all epochs history
hist=pickle.load(open('Y:/lab2/ShowTell/hist0.p', 'rb'))
hist1=pickle.load(open('Y:/lab2/ShowTell/hist1.p', 'rb'))
hist2=pickle.load(open('Y:/lab2/ShowTell/hist2.p', 'rb'))
hist3=pickle.load(open('Y:/lab2/ShowTell/hist3.p', 'rb'))
hist4=pickle.load(open('Y:/lab2/ShowTell/hist4.p', 'rb'))
hist5=pickle.load(open('Y:/lab2/ShowTell/hist5.p', 'rb'))
hist6=pickle.load(open('Y:/lab2/ShowTell/hist6.p', 'rb'))
hist7=pickle.load(open('Y:/lab2/ShowTell/hist7.p', 'rb'))
hist8=pickle.load(open('Y:/lab2/ShowTell/hist8.p', 'rb'))
hist9=pickle.load(open('Y:/lab2/ShowTell/hist9.p', 'rb'))

###convert to single list
history.append(hist['loss'])
history.append(hist1['loss'])
history.append(hist2['loss'])
history.append(hist3['loss'])
history.append(hist4['loss'])
history.append(hist5['loss'])
history.append(hist6['loss'])
history.append(hist7['loss'])
history.append(hist8['loss'])
history.append(hist9['loss'])

```

```python
###plot 
plt.title('Training Loss : categorical_crossentropy ')
plt.plot(history,label='train ')
plt.legend(loc='lower right')
plt.xlabel('epoch')
plt.show()
```
